```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.sample(5)
```

```python
titanic = pd.read_csv('train.csv')
```

```python
titanic.sample(5)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.sample(5)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.sample(5)
```

<h2>0. Lineplot (Numerical - Numerical)</h2>

```python
numeric_columns = flights.select_dtypes(include='number').columns
new = flights.groupby('year')[numeric_columns].sum()
sns.lineplot(data = new, x = 'year', y = 'passengers', label = 'Per Year Travelors', color = 'aqua', linewidth = 2.44)
```

<h2>1. Scatterplot (Numerical - Numerical)</h2>

```python
sns.scatterplot(data =  tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker', size = 'size')
plt.show()
tips['smoker'].value_counts()
```

<h2>2. Barplot (Numerical - Categorical)</h2>

```python
sns.barplot(data = titanic, x = 'Pclass', y = 'Age')
plt.show()
```

```python
# Fare in dollors:
sns.barplot(data = titanic, x = 'Pclass', y = 'Fare', hue = 'Sex')
plt.show()
```

<h2>3. Boxplot (Numerical - Categorical)</h2>

```python
sns.boxplot(data = titanic, x = 'Sex', y = 'Age', hue = 'Survived')
plt.show()
```

<h2>4. kdeplot (Numerical - Categorical)</h2>

```python
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'],color='darkblue' , label='Not Survived')
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'], color='orangered', label='Survived')
plt.show()
```

<h2>5. Heatmap (Categorical - Categorical)</h2>

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
plt.show()
```

```python
flights.pivot_table(values='passengers', index='month', columns='year', observed=True)
sns.heatmap(flights.pivot_table(values='passengers', index='year', columns='month', observed=True))
```

```python
# Clustermap of the above data to see the similar behaving months:
flights.pivot_table(values='passengers', index='month', columns='year', observed=True)
sns.clustermap(flights.pivot_table(values='passengers', index='year', columns='month', observed=True))
```
