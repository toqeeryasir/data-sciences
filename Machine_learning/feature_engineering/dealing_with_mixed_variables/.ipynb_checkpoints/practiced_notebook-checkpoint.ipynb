```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.sample(10)
```

```python
df.rename(columns={"number": "Fmembers"}, inplace=True)
```

```python
df.head()
```

# Dealing with Cabin column:

```python
# how to drop a column

# df = df.drop(columns= "Survived")
# del df["Survived"]
# removed_column = df.pop("Survived")
# df = df.drop(columns=[df.columns[-1]])
df.drop(columns=[df.columns[-1]], inplace=True)
```

```python
df.head()
```

```python
df["Cabin_Name"] = df["Cabin"].str.extract(r"([A-Za-z]+)", expand=False)  # Extract full letters
df["Berth_Number"] = df["Cabin"].str.extract(r"(\d+)", expand=False)  # Extract numbers

# Correct numeric conversion (assigning back to column)
df["Berth_Number"] = pd.to_numeric(df["Berth_Number"], errors="coerce").astype("Int64")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns="Cabin", inplace=True)
```

```python
df.head()
```

# Dealing with Fmembers column:

```python
df["Fmembers_Alphas"] = df["Fmembers"].str.extract(r"([A-Za-z]+)", expand=False)
df["Fmembers_Int"] = df["Fmembers"].str.extract(r"(\d+)", expand=False)

df["Fmembers_Int"] = pd.to_numeric(df["Fmembers_Int"], errors='coerce').astype("Int64")
```
