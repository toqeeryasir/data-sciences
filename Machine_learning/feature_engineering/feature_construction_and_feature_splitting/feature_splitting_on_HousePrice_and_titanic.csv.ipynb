```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['Name'].str.split(', ', expand=True)
```

```python
df['SirName'] = df['Name'].str.split(', ', expand=True)[0]
```

```python
df.head()
```

```python
df['Name'].str.split(', ', expand=True)[1].str.split('. ', expand = True)[0]
```

```python
df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand = True)[0]
df.head()
```

```python
df['FullName'] = df['Name'].str.split(', ', expand=True)[0] +' '+ df['Name'].str.split(', ', expand=True)[1].str.split('. ', expand = True)[2]
```

```python
df.head()
```

```python
df['LastName'] = df['Name'].str.split(', ', expand=True)[1].str.split('. ', expand = True)[2]
```

```python
df.head()
```

```python
new_order = [
    'PassengerId', 'Name', 'FullName', 'SirName', 'LastName', 'Title', 
    'Sex', 'Age', 'Pclass', 'SibSp', 'Parch', 
    'Ticket', 'Fare', 'Cabin', 'Embarked', 'Survived'
]

df = df[new_order]
```
