```python
import numpy as np
import pandas as pd
from datetime import date
```

```python
df = pd.read_csv('house_price.csv')
```

```python
df.head()
```

```python
variable_names = []
for variable in df.columns:
    variable_names.append(variable)
variable_names
```

```python
current_date = date.today().year
current_date
```

```python
# adding new feature HouseAge:
df['HouseAge'] = current_date - df['YearBuilt']
```

```python
print(df['HouseAge'].max())
print(df['HouseAge'].min())
```

```python
# adding new feature Total baths:
df['TotalBathrooms'] = df['FullBath'] + df['HalfBath'] + df['BsmtFullBath'] + df['BsmtHalfBath']
```

```python
# adding new feature TotalSquareFootage:
df['TotalSF'] = df['GrLivArea'] + df['TotalBsmtSF']
```
