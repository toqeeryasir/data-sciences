```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("house_price.csv", usecols=["GarageQual", "FireplaceQu", "SalePrice"])
```

```python
df.head()
```

```python
# missing values:
df.isnull().mean()*100
```

```python
sns.barplot(df["GarageQual"].value_counts().sort_values(ascending=False))
plt.show()
```

```python
df["GarageQual"].mode()
```

```python
TA_temp = df[df["GarageQual"]=="TA"]
```

```python
sns.kdeplot(x = df[df["GarageQual"]== "TA"]["SalePrice"], color="green", label = "TA values")
sns.kdeplot(x = df[df["GarageQual"].isnull()]["SalePrice"], color="red", label = "Missing values")
plt.legend()
plt.show()
```

```python
df["GarageQual"].fillna("TA", inplace=True)
```

```python
sns.barplot(df["GarageQual"].value_counts())
```

```python
sns.kdeplot(x = df[df["GarageQual"]== "TA"]["SalePrice"], color="green", label = "TA values")
sns.kdeplot(x = TA_temp["SalePrice"], color="red", label = "Missing values", alpha = 0.8)
plt.legend()
plt.show()
```

```python
df["FireplaceQu"].mode()
```

```python
Gd_temp = df[df["FireplaceQu"]=="Gd"]
```

```python
sns.barplot(df["FireplaceQu"].value_counts().sort_values(ascending=False))
plt.show()
```

```python
df["FireplaceQu"].fillna("Gd", inplace=True)
```

```python
sns.kdeplot(x = df[df["FireplaceQu"]== "Gd"]["SalePrice"], color="green", label = "Gd values")
sns.kdeplot(x = Gd_temp["SalePrice"], color="red", label = "Missing values", alpha = 0.7)
plt.legend()
plt.show()
```
