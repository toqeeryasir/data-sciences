```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
```

```python
covid= pd.read_csv("covid_19_pakistan.csv")
```

```python
covid.sample(6)
```

```python
covid.info()
```

```python
sns.displot(data = covid, x = "Cases", bins = 20, kde = True, stat = "probability")
plt.show()
covid["Cases"].value_counts()
```

```python
sns.scatterplot(data = covid, x = "Deaths", y = "Recovered", hue = "Cases", style = "Province")
```

```python
sns.lineplot(data = covid, x = "Deaths", y = "Recovered", color = "red")
```

```python
sns.histplot(data = covid, x = "Deaths")
```

```python
sns.histplot(data = covid, x = "Recovered")
```

```python
sns.lineplot(data = covid, x = "Cases", y = "Recovered", color = "orange")
```

```python
sns.lineplot(data = covid, x = "Cases", y = "Deaths", color = "aqua")
```
