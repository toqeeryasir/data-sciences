<h1>Choosing the Right Plot:</h1>

```python
#Distribution of a single numerical column: Use histograms or boxplots.
#Comparison between two numerical columns: Use scatter plots or line plots.
#Correlation among multiple numerical columns: Use a heatmap.
```

```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df.sample(5)
```

```python
# methode to plot a series of data insted of data_freme:
sns.countplot(x=df['Survived'])
plt.show()
```

```python
# methode to plot a dara_frame:
sns.countplot(x= 'Survived', data = df)
plt.show()
df['Survived'].value_counts()
```

```python
#second methode to do same:
df['Survived'].value_counts().plot(kind = 'bar')
```

```python
sns.countplot(x = 'Sex', data = df)
```

```python
sns.countplot(x = 'Embarked', data = df)
plt.show()
```

```python
sns.countplot(x = 'Pclass', data = df)
plt.show()
df['Pclass'].value_counts()
```

```python
sns.displot(data = df, x='Age', kde=True, bins=10, stat="probability")
plt.show()
```

```python
values = df['Survived'].value_counts()
plt.pie(values, labels=values.index, autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
values = df['Sex'].value_counts()
plt.pie(values, labels = values.index, autopct = '%1.1f%%', startangle = 90)
plt.show()
```

```python
values = df['Embarked'].value_counts()
plt.pie(values, labels = values.index, autopct ='%1.1f%%', startangle = 90)
plt.show()
```

```python
values = df['Pclass'].value_counts()
plt.pie(values, labels = values.index, autopct = '%1.1f%%', startangle = 90)
plt.show()
```

```python
sns.histplot(df['Age'], bins = 8, kde = False)
plt.show()
df['Age'].value_counts()
```

```python
sns.histplot(df['Age'], bins = 8, kde = True)
plt.show()
```

```python
#Numerical vs categorical data:
sns.boxplot(data = df, x = 'Age')
```

```python
sns.boxplot(data = df, x = 'Fare')
```

```python
# Minimum age of a passenger:
df['Age'].min()
```

```python
# Maximum age of a passenger:
df['Age'].max()
```

```python
# Mean age of passenger:
df['Age'].mean()
```

```python
#Numerical vs numerical data:
sns.scatterplot(data = df, x = 'Age', y = 'Fare')
plt.show()
```

<h1>To convert data into a csv file:</h1>

```python
pdf = df['Age'].value_counts()
pdf_data = pdf.reset_index()
pdf_data.columns = ['Age', 'count']
pdf_data.to_csv('age_counts.csv', index = False)
```
