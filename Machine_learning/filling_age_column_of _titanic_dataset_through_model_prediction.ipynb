```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("titanic.csv")
```

```python
data2 = data1 = data[["Age", "Fare"]]
```

```python
data1.head()
```

```python
data2.isnull().sum()
```

```python
data2.info()
```

```python
data2 = data2.dropna()
```

```python
data2.info()
```

```python
data2.isnull().sum()
```

<h1>Training A Data Model To Fill Missing Values Based On Prediction</h1>

```python
x_train = data2[["Fare"]]
y_train = data2["Age"]
```

```python
age_predictor =LinearRegression()
age_predictor.fit(x_train, y_train)
```

```python
missing_data2 = missing_data = data1[data1["Age"].isnull()]
```

```python
missing_data.head()
```

```python
missing_data["Age"].isnull().sum()
```

```python
missing_data.count()
```

```python
x_missing = missing_data[["Fare"]]
```

```python
x_missing.head()
```

```python
predicted_age = age_predictor.predict(x_missing)
```

<h1>Filled Age Col By Prediction</h1>

```python
data1.loc[data["Age"].isnull(), "Age"] = predicted_age
```
